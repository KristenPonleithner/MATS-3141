```python
#the difference between the numerical and analytical ratio was 0.037. 
import random
count = 0
for i in range(100000):
    a = random.randint(1,6)
    b = random.randint(1,6)
    if a + b == 7:
        count += 1
ratio = count/100000
print(ratio)
```

Analytical solution:

The combinations for two dice that add up to 7 are [1,6], [2,5], and [3,4].
The total possible combinations are $\frac{6*6}{2!}=18$ (divide by 2! because two dice give 2 permutations per unique combination.)

The probability of rolling a pair that sums to 7 is $\frac{3}{18}=\frac{1}{6}$
